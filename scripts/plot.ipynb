```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
countries_aggregated = pd.read_csv('../data/countries-aggregated.csv')
```

```python
#function to process data
def preprocess(data,countries):
    country_data = {}
    for country in countries:
        df = data.loc[data['Country']==country]
        df = pd.DataFrame(df)
        df.reset_index(inplace=True)
        country_data[country] = df
    return country_data
```

```python
#function to plot date against [Confirmed,Recovered,Deaths]
def plotdata(data):
    for country in data:
        confirmed = data[country].Confirmed
        recovered = data[country].Recovered
        death = data[country].Deaths
        dates = data[country].Date
        plt.figure(figsize=(20,10))
        plt.title('COVID %s'%country,fontsize=20)
        plt.xlabel('Dates',fontsize=18)
        plt.ylabel('Numbers',fontsize=18)
        plt.xticks(rotation = 45)
        plt.plot(dates,confirmed)
        plt.plot(dates,recovered)
        plt.plot(dates,death)
```

```python
countries = ['China','India','US','United Kingdom','Italy']
```

```python
country_data = preprocess(countries_aggregated, countries)
```

```python
plotdata(country_data)
```
